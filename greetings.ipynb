```python
def greeting_response(message):

    response = ""
    greetings = ["hi", "hello", "hey"]

    if any(word in message.lower() for word in greetings):
        response = "Hi!"
    else:
       response = "All is well"

    return response
    pass

# Test cases
test_messages = [
    "Hello there!",
    "How's it going?"
]

for msg in test_messages:
    response = greeting_response(msg)
    print(f"You: {msg}")
    print(f"Bot: {response}")
```
